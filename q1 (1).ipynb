```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
given_beta1 = 262
given_beta2 = -0.006
given_beta3 = -2.4
given_var = 42
```

## generating 1000 sets of 64 observations of ui

```python
import math
u=[]
for i in range(1000):
    u.append([np.random.randn()*math.sqrt(42) for i in range(64)])
u
```

## Calculating beta's

```python
df = pd.read_csv("Table3.csv")
df
```

```python
del df['TFR'] 
# because TFR is of no use in regression
```

```python
df
```

```python
reg = linear_model.LinearRegression()

y_CM=df['CM']
y
Y=[[0 for i in range(64)] for i in range(1000)]
for j in range(len(u)):
    for i in range(len(y_CM)):
        Y[j][i]=y_CM[i]-u[j][i]

beta_list=[]        
for i in range(len(Y)):
    reg.fit(df[['FLR','PGNP']],Y[i])
    beta_sublist=[reg.intercept_,reg.coef_[1],reg.coef_[0]]
    beta_list.append(beta_sublist)


betas=[]
for j in range(3):
    a=[]
    for i in range(len(beta_list)):
        a.append(beta_list[i][j])
    betas.append(sum(a)/len(beta_list))
B1,B2,B3 = betas
print(B1,B2,B3)
```

```python
print((B1-given_beta1)/given_beta1)
print((B2-given_beta2)/given_beta2)
print((B3-given_beta3)/given_beta3)
```
